```python
from skimage import data
import numpy as np
import matplotlib.pyplot as plt

caller = getattr(data, 'camera')
image = caller()

np.save('camera512.npy',image)

image256 = image[0::2,0::2]
np.save('camera256.npy',image256)

image128 = image256[0::2,0::2]
np.save('camera128.npy',image128)

fig, ax = plt.subplots(dpi=150)
ax.imshow(image128)


caller = getattr(data, 'gravel')
image = caller()

np.save('gravel512.npy',image)

image256 = image[0::2,0::2]
np.save('gravel256.npy',image256)


image128 = image256[0::2,0::2]
np.save('gravel128.npy',image128)

fig, ax = plt.subplots(dpi=150)
ax.imshow(image128)
```
